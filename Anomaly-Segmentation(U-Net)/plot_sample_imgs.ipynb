```python
from glob import glob
import matplotlib.pyplot as plt
import os
import numpy as np
import matplotlib.gridspec as gridspec



%matplotlib inline
```

```python
def plot_n_over_n_grid_of_samples(path, n):
    subfolders = np.random.choice(glob(path+"*"), n*n, replace=False)

    plt.figure(figsize=(18, 18))
    gs1 = gridspec.GridSpec(n, n)
    gs1.update(wspace=0.001, hspace=0.001) # set the spacing between axes. 

    for i, subfolder in enumerate(subfolders):
        img = plt.imread(glob(subfolders[i]+"\\images\\*")[0])
        plt.subplot(n, n, i+1)
        plt.axis('off')
        plt.title(glob(subfolders[i]+"\\images\\*")[0].split("\\")[-1][-10:])
        plt.imshow(img[1:])
```

```python
path = "stage1_train/"
plot_n_over_n_grid_of_samples(path, 5)
```

```python
path = "stage1_test/"
plot_n_over_n_grid_of_samples(path, 5)
```

```python
path = "stage2/"
plot_n_over_n_grid_of_samples(path, 5)
```
